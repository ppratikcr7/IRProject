```python
#centroids = np.array((random.sample(list(data),3)))
centroids = np.random.rand(3,4)
centroids
```

```python
from scipy.spatial import distance

dist = distance.cdist(data,centroids,'euclidean')

dist
```

```python
MemberShip = []

for i in range(len(dist)):
    
    if dist[i][0]==0:
        dist[i][0] = dist[i][0] + 0.01
        MemberShip = np.append(MemberShip,(1/dist[i][0])/1/np.sum(dist[i]))
    else:
        MemberShip = np.append(MemberShip,(1/dist[i][0])/1/np.sum(dist[i]))
        
    if dist[i][1]==0:
        dist[i][1] = dist[i][1] + 0.01
        MemberShip = np.append(MemberShip,(1/dist[i][1])/1/np.sum(dist[i]))
    else:
        MemberShip = np.append(MemberShip,(1/dist[i][1])/1/np.sum(dist[i]))
        
        
    if dist[i][2]==0:
        dist[i][2] = dist[i][2] + 0.01
        MemberShip = np.append(MemberShip,(1/dist[i][2])/1/np.sum(dist[i]))
    else:
        MemberShip = np.append(MemberShip,(1/dist[i][2])/1/np.sum(dist[i]))
```

```python
MemberShip = np.reshape(MemberShip,(-3,3))
```

```python
centroid = []
for i in range(len(centroids)):
    centroid = np.append(centroid,np.matmul(np.transpose(MemberShip[:,i]),data)/sum(MemberShip[:,i]),axis=0)


print(centroids)
np.reshape(centroid,(-4,4))
```
